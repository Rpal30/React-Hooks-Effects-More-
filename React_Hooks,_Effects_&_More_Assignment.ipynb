Question 1 : In simple words, what are React Hooks? Create a small component that
shows a button. When the button is clicked, toggle the text between “On” and “Off".

React Hooks are special functions in React that let you “hook into” React features like state and lifecycle methods without writing a class. They make it easier to manage things like data, side effects, and component logic in functional components.

UseState → lets you add state (data that changes over time).

UseEffect → lets you run code when something changes (like fetching data or updating the DOM).

Example: Toggle Button Component

```python
import React, { useState } from "react";

function ToggleButton() {
  // Declare a state variable called "isOn"
  const [isOn, setIsOn] = useState(false);

  // Function to toggle the state
  const handleClick = () => {
    setIsOn(!isOn);
  };

  return (
    <div>
      <p>{isOn ? "On" : "Off"}</p>
      <button onClick={handleClick}>Toggle</button>
    </div>
  );
}

export default ToggleButton;
```

This component:

Starts with isOn = false (so it shows “Off”).

When you click the button, it flips the state.

The text updates instantly to “On” or “Off”.

Question 2 : Make a counter with a button. Whenever the count changes, log “Count
changed” using useEffect.


Counter with useEffect :

useState → to store and update the counter value.

useEffect → to run side effects (like logging) whenever the count changes.

Example Code :

```python
import React, { useState, useEffect } from "react";

function Counter() {
  // State to keep track of count
  const [count, setCount] = useState(0);

  // useEffect runs whenever 'count' changes
  useEffect(() => {
    console.log("Count changed");
  }, [count]); // dependency array ensures effect runs only when count changes

  return (
    <div>
      <p>Count: {count}</p>
      <button onClick={() => setCount(count + 1)}>Increase</button>
    </div>
  );
}

export default Counter;
```

How it Works :

The state variable count starts at 0.

Clicking the button increases the count by 1.

The effect runs every time count changes, logging “Count changed” in the console.

Question 3 : Write a component that fetches users from
https://jsonplaceholder.typicode.com/users only once on mount and
shows how many users were loaded.

Fetching Users with useEffect
We’ll build a React component that fetches user data from https://jsonplaceholder.typicode.com/users only once when the component mounts, and then shows how many users were loaded.

useState → to store the list of users.

useEffect → to fetch data only once when the component mounts.

Example Code :

```python
import React, { useState, useEffect } from "react";

function UserLoader() {
  const [users, setUsers] = useState([]);

  useEffect(() => {
    // Fetch users only once on mount
    fetch("https://jsonplaceholder.typicode.com/users")
      .then((response) => response.json())
      .then((data) => setUsers(data))
      .catch((error) => console.error("Error fetching users:", error));
  }, []); // empty dependency array ensures it runs only once

  return (
    <div>
      <h2>Users Loaded: {users.length}</h2>
    </div>
  );
}

export default UserLoader;
```

How it Works :

The state users starts as an empty array.

On mount, the effect runs once (because of the empty dependency array []).

It fetches data from the API, converts it to JSON, and updates the state.

The component then displays the number of users loaded.

Question 4 : Create a component that starts a timer with setInterval to increase a
“seconds” counter every second. Stop the timer when the component unmount.

Example Code :

```python
import React, { useState, useEffect } from "react";

function Timer() {
  const [seconds, setSeconds] = useState(0);

  useEffect(() => {
    // Start the interval
    const interval = setInterval(() => {
      setSeconds((prevSeconds) => prevSeconds + 1);
    }, 1000);

    // Cleanup: stop the interval when component unmounts
    return () => {
      clearInterval(interval);
    };
  }, []); // empty dependency array ensures it runs only once

  return (
    <div>
      <h2>Seconds: {seconds}</h2>
    </div>
  );
}

export default Timer;
```

How it Works :

The state seconds starts at 0.

On mount, the effect sets up a setInterval that increases seconds every 1000 ms.

When the component unmounts, the cleanup function runs and clears the interval, preventing memory leaks or unwanted background timers.

Question 5 : Build a small form with a text input. If the input is empty, show “Please
type your name”. If it has text, show “Hello, <name>”.

Example Code :

```python
import React, { useState } from "react";

function NameForm() {
  const [name, setName] = useState("");

  return (
    <div>
      <input
        type="text"
        placeholder="Enter your name"
        value={name}
        onChange={(e) => setName(e.target.value)}
      />
      <p>
        {name === "" ? "Please type your name" : `Hello, ${name}`}
      </p>
    </div>
  );
}

export default NameForm;
```

How it Works :

The useState hook stores the input value (name).

The input field updates the state whenever the user types.

The p element conditionally renders:

If name is empty → shows “Please type your name”.

Otherwise → shows “Hello, <name>”.

Question 6 : You have a theme value (“light” or “dark”) that is needed by a deep child
button. Show how to pass it without sending theme through every parent prop.


Passing Theme Without Prop Drilling :

In React, when you need to share a value (like a theme: "light" or "dark") across deeply nested components without passing it through every parent prop, you use the Context API.

Example Code :

```python
import React, { createContext, useContext } from "react";

// 1. Create a Context
const ThemeContext = createContext("light");

function App() {
  return (
    // 2. Provide the theme value at the top level
    <ThemeContext.Provider value="dark">
      <Toolbar />
    </ThemeContext.Provider>
  );
}

function Toolbar() {
  return (
    <div>
      <DeepChildButton />
    </div>
  );
}

function DeepChildButton() {
  // 3. Consume the theme value directly
  const theme = useContext(ThemeContext);

  return <button>Current theme: {theme}</button>;
}

export default App;
```

How it Works :

createContext → creates a context object.

Provider → wraps components and provides the theme value.

useContext → lets the deep child button access the theme directly, without props.

Question 7 : Make a counter with two buttons: “+1” and “-1”. Use useReducer instead
of useState.


Instead of using useState, we can manage state with useReducer. This is useful when state logic gets more complex, but even for a simple counter it shows how reducers work.

Example Code :

```python
import React, { useReducer } from "react";

// Reducer function: decides how state changes based on action
function counterReducer(state, action) {
  switch (action.type) {
    case "increment":
      return state + 1;
    case "decrement":
      return state - 1;
    default:
      return state;
  }
}

function Counter() {
  // useReducer returns [state, dispatch]
  const [count, dispatch] = useReducer(counterReducer, 0);

  return (
    <div>
      <p>Count: {count}</p>
      <button onClick={() => dispatch({ type: "increment" })}>+1</button>
      <button onClick={() => dispatch({ type: "decrement" })}>-1</button>
    </div>
  );
}

export default Counter;
```

How it Works :

The reducer function (counterReducer) defines how the state changes depending on the action type.

useReducer gives us the current state (count) and a dispatch function to send actions.

Clicking +1 dispatches an "increment" action, while -1 dispatches a "decrement" action.

The reducer updates the state accordingly.

Question 8 : Given an array nums = [1,2,3,4] and a filter value min, show how to
compute a filtered list only when nums or min change. Also pass a stable click handler
to a child list item.

Filtering with useMemo and Stable Handlers with useCallback :

When you want to compute a filtered list only when dependencies change (like nums or min), you use useMemo.

When you want to pass a stable click handler to child components (so it doesn’t get recreated on every render), you use useCallback.

Example Code :

```python
import React, { useState, useMemo, useCallback } from "react";

function App() {
  const [nums] = useState([1, 2, 3, 4]);
  const [min, setMin] = useState(2);

  // Compute filtered list only when nums or min change
  const filteredList = useMemo(() => {
    return nums.filter((n) => n >= min);
  }, [nums, min]);

  // Stable click handler
  const handleClick = useCallback((num) => {
    console.log("Clicked:", num);
  }, []);

  return (
    <div>
      <h2>Filtered List (min = {min}):</h2>
      <ul>
        {filteredList.map((num) => (
          <ListItem key={num} num={num} onClick={handleClick} />
        ))}
      </ul>
      <button onClick={() => setMin(min + 1)}>Increase Min</button>
    </div>
  );
}

function ListItem({ num, onClick }) {
  return <li onClick={() => onClick(num)}>{num}</li>;
}

export default App;
```

How it Works :

useMemo → ensures the filtered list is recalculated only when nums or min change.

useCallback → ensures the click handler reference stays stable across renders, preventing unnecessary re-renders of child components.

The child ListItem receives a stable onClick function, so React doesn’t think it’s a “new” prop every time.

Question 9 : Create a controlled input for email. As the user types, show “Email:
value”. Why do we keep the input’s value in state?


Controlled Email Input in React :

A controlled input means the input’s value is managed by React state. This way, React always knows what’s inside the input field, and the UI stays in sync with the state.

Example Code :

```python
import React, { useState } from "react";

function EmailInput() {
  const [email, setEmail] = useState("");

  return (
    <div>
      <input
        type="email"
        placeholder="Enter your email"
        value={email}
        onChange={(e) => setEmail(e.target.value)}
      />
      <p>Email: {email}</p>
    </div>
  );
}

export default EmailInput;
```

Controlled inputs → ensure React is the “single source of truth” for the form data.

It allows validation, formatting, and conditional rendering based on the input value.

Makes it easy to reset, prefill, or submit the data consistently.

Without state, the input would be uncontrolled, meaning React wouldn’t know its value until you manually read it from the DOM.

Question 10 : Make a button that focuses an input using a ref. Return adjacent
elements without extra wrapper nodes. In one sentence, say what React’s update
process does.


Example Code :

```python
import React, { useRef } from "react";

function FocusInput() {
  const inputRef = useRef(null);

  const handleFocus = () => {
    inputRef.current.focus();
  };

  return (
    <>
      <input type="text" ref={inputRef} placeholder="Type here..." />
      <button onClick={handleFocus}>Focus Input</button>
    </>
  );
}

export default FocusInput;
```

How it Works :

useRef → creates a reference to the input element.

inputRef.current.focus() → focuses the input when the button is clicked.

Fragments (<> ... </>) → allow returning multiple adjacent elements without adding extra <div> wrappers.
